```python
import requests
import numpy as np
import pandas as pd
import json
headers={'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/70.0.3538.110 Safari/537.36'}
dd=pd.DataFrame()
def xiaohonsghu_text(r):
    text=json.loads(r.text)
    bb=pd.DataFrame()
    for i in  np.arange(len(text['data'])):
        aa=pd.DataFrame()
        aa['商品编号']=[text['data'][i]['id']]
        aa['商家编号']=[text['data'][i]['seller_id']]
        aa['商品图片']=[text['data'][i]['image']]
        aa['文章标题']=[text['data'][i]['title']]
        aa['商品描述']=[text['data'][i]['desc']]
        aa['商品现价']=[text['data'][i]['discount_price']]
        aa['商品链接']=[text['data'][i]['link']]
        try:
            aa['活动名称']=[text['data'][i]['promotion_text']]
        except:
            pass
        aa['店铺名称']=[text['data'][i]['vendor_name']]
        aa['店铺链接']=[text['data'][i]['vendor_link']]
        try:
            aa['商品原价']=[text['data'][i]['price']]
        except:
            pass
        try:
            aa['会员价']=[text['data'][i]['member_price']]
        except:
            pass
        aa['商品特点']=[text['data'][i]['feature']]
        aa['是否包税']=[text['data'][i]['tax_included']]
        aa['是否新品']=[text['data'][i]['new_arriving']]
        aa['是否可被购买']=[text['data'][i]['buyable']]  
        bb=bb.append(aa)

    cc=bb[['商品编号','文章标题','商品描述','商品特点','是否包税','是否新品','是否可被购买','商品链接','商品现价','商品原价','会员价','活动名称','商品图片','商家编号','店铺名称','店铺链接']]
    #cc.to_csv(r'/Users/shana/Documents/xiaohongshu.csv')
    return(cc)
```

```python
for i in np.arange(3)+1:
    link='https://www.xiaohongshu.com/api/store/vs/53df5710b4c4d6383ae8e9a6/items/v2?page={}&sort=sales_qty'.format(i)
    r=requests.get(link,headers=headers,timeout=10)
    cc=xiaohonsghu_text(r)
    if i==1:
        dd=cc
    else:
        dd=dd.append(cc)
dd.to_csv(r'/Users/shana/Documents/xiaohongshu.csv')
```
